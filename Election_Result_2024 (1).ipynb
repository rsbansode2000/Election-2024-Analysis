##**Visualizing Data**



*   **Party Wise Distribution**




```
party_wise = df.groupby(by=['winning_party'], as_index = False).size().sort_values(by='size', ascending = False).reset_index(drop=True)
party_wise
```

```
party_wise['seats in %'] = round(party_wise['size']*100/party_wise['size'].sum(), 2)
party_wise
```

#Required Dataset for pie chart




```
required_result = 5

required_party_wise = party_wise.head(required_result)
other_size = int(party_wise.loc[required_result:].agg({'size': 'sum'}))
other_percentage = float(100-required_party_wise.agg({'seats in %' : 'sum'}))

df1 = {'winning_party': 'Other', 'size': other_size, 'seats in %': other_percentage}
required_party_wise = required_party_wise._append(df1, ignore_index = True)
```

```
required_party_wise
```

# **Party Wise Seat Distribution**

```
colors = ['orange', 'green','blue', 'yellow', 'red', '#e2e2e2']
plt.figure(figsize = (10,15))

plt.pie(required_party_wise['size'], colors= colors, labels = required_party_wise['size'].astype('str')+' seats', radius = 0.8,
        autopct = '%.2f %%',wedgeprops={'edgecolor' : 'black', 'linewidth' : 1}
        ,explode=[0.005,0.01,0.01,0.01,0.01,0.01],
        pctdistance= 0.7, labeldistance = 1.1)

#Adding circle at center to make donut shape
# plt.gca().add_artist(plt.Circle((0,0), radius = 0.5, fc = 'white'))

plt.xlabel('Number of seats won by Party')
plt.title('Lok Sabha 2024 Party Wise Seat Distribution')
plt.legend(loc = 'upper right', labels = required_party_wise['winning_party'])

plt.show()
```

From above chart we can say that BJP is close to get 50% seats.

# **Top Parties**

```
total_seats_per_state = df.groupby('State', as_index = False).size().sort_values('size', ascending = False)
total_seats_per_state
```

```
bjp = df[df['winning_party'].str.contains('Bharatiya Janata Party')].groupby('State', as_index = False).size().sort_values('size', ascending = False)
inc = df[df['winning_party'].str.contains('Indian National Congress')].groupby('State', as_index = False).size().sort_values('size', ascending = False)
samajwadi = df[df['winning_party'].str.contains('Samajwadi Party')].groupby('State', as_index = False).size().sort_values('size')
trinamol_congress =df[df['winning_party'].str.contains('All India Trinamool Congress')].groupby('State', as_index = False).size().sort_values('size')
```

```
r1 =pd.merge(total_seats_per_state, bjp, on='State', how='left').rename(columns = {'size_x' : 'total_seats', 'size_y' : 'bjp'})
r1.fillna(0, inplace = True)
r1['bjp'] = r1['bjp'].astype('int')
r2 =pd.merge(r1, inc, on='State', how='left').sort_values(by=['total_seats', 'bjp'], ascending = False).rename(columns = {'size' : 'inc'})
r2.fillna(0, inplace = True)
r2['inc'] = r2['inc'].astype('int')
r3 =pd.merge(r2, samajwadi, on='State', how='left').sort_values(by=['total_seats', 'bjp'], ascending = False).rename(columns = {'size' : 'samajwadi'})
r3['samajwadi'].fillna(0, inplace = True)
r3['samajwadi'] = r3['samajwadi'].astype('int')
r4 =pd.merge(r3, trinamol_congress, on='State', how='left').sort_values(by=['total_seats', 'bjp'], ascending = False).rename(columns = {'size' : 'trinamol_congress'})
r4['trinamol_congress'].fillna(0, inplace = True)
r4['trinamol_congress'] =r4['trinamol_congress'].astype('int')
big_parties = r4.copy()
```

```
big_parties
```

```
plt.figure(figsize=(30,15))
N = 36
my_index = np.arange(N)
width = 0.3
b = plt.bar(big_parties['State'].str.slice(0,4), big_parties['total_seats'],width = width*2, color = '#e2e2e2')


b1 = plt.bar(my_index-width/2, big_parties['bjp'], width, color = 'orange')
b2 = plt.bar(my_index+width/2, big_parties['inc'], width, color = 'blue')


# plt.xticks(ind+width*3)
plt.legend( (b ,b1, b2, b3), ('Total','BJP', 'Congress') , title = 'Top Party', fontsize= 20, title_fontsize = 30, facecolor = '#e6f9ff')
plt.grid(axis = 'y', color = 'green', linestyle = '--', linewidth = 1)

plt.xlabel('State', fontsize = 20,fontdict={'fontweight' : 'bold'})
plt.ylabel('Seats won', fontsize = 20, fontdict={'fontweight' : 'bold'})
plt.title('State Wise Seat', fontdict={'fontweight' : 'bold', 'fontsize' : 30})

plt.show()
```
